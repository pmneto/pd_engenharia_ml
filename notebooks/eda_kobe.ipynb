# Análise Exploratória dos Dados (EDA)

Este notebook tem como objetivo explorar os dados utilizados no projeto de predição de arremessos do Kobe Bryant.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Carregando os dados já processados
df = pd.read_parquet("../data/processed/data_filtered.parquet")
df.shape
```

```python
# Checando valores nulos
df.isnull().sum()
```

```python
# Distribuição da variável alvo
sns.countplot(x="shot_made_flag", data=df)
plt.title("Distribuição de Acertos e Erros")
plt.show()
```

```python
# Matriz de correlação entre variáveis numéricas
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="Blues")
plt.title("Matriz de Correlação")
plt.show()
```

```python
# Scatterplot para localização dos arremessos
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="lng", y="lat", hue="shot_made_flag", alpha=0.6)
plt.title("Distribuição dos Arremessos por Localização")
plt.xlabel("Longitude (lng)")
plt.ylabel("Latitude (lat)")
plt.legend(title="Acertou?")
plt.show()
```
